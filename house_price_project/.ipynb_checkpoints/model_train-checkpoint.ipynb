```python
import pandas as pd

# Load the house dataset
df = pd.read_csv(r"C:\Users\msjoh\Downloads\Housing.csv")
df
```

```python
# Information for all columns
df.info()
```

```python
# Statistical Descriptions
df.describe()
```

```python
# Find any null values
df.isnull().sum()
```

```python
# Value counts for each column
df.nunique()
```

```python
# Encoding the categorical columns
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

cols = ["mainroad","guestroom","basement","hotwaterheating","airconditioning","prefarea","furnishingstatus"]

for col in cols:
    df[col] = le.fit_transform(df[col])

df
```

```python
# Train-Test Split
x = df.drop("price", axis=1) # => Traning data
y = df["price"] # => Test data
```

```python
# Split into 80:20 format
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2, random_state=42)
```

```python
# Feature Scaling 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# Train the Model
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()

lr_model.fit(x_train, y_train)
```

```python
# Predict the value
y_pred = lr_model.predict(x_test)
```

```python
# Find the Accuracy
from sklearn.metrics import r2_score

accuracy = r2_score(y_test, y_pred)
accuracy
```

```python
# Save the model and scaler
import joblib

joblib.dump(lr_model, "house_model.pkl")
joblib.dump(scaler, "scaler.pkl")
```
